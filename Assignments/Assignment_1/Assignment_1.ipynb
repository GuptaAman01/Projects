```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading a CSV file

data=pd.read_csv('Credit.csv')
data
```

```python
# Droppping locality column because this column contains a lot ofmissing values

data=data.drop(["Locality"], axis=1)
```

```python
# Dropping Employment column becuase this column contains a lot of missing values

data=data.drop(["Employment"], axis=1)
```

```python
data
```

```python
# Duplicate value in data frame

data_duplicate=data[data.duplicated()]
data_duplicate
```

```python
# Dropping duplicate value

data=data.drop_duplicates(keep=False)
```

```python
# Removing missing values from Age column

data=data.drop(data["Age"].index[data["Age"].isna()])
data
```

```python
# Removing nan values from Age column

data=data.drop(data["Age"].index[data["Age"]=="nan"])
```

```python
# Removing special characters from Age column

data["Age"]=data["Age"].astype(str)
data["Age"]=data.Age.apply(lambda x : int(x[:-1]) if x[-1]=='%' else int(x))
data["Age"]=np.where((data["Age"]>100),100,data["Age"])
data["Age"]
```

```python
# Max value of Age

val_max=data["Age"].max()
val_max
```

```python
# Min value of Age

val_min=data["Age"].min()
val_min
```

```python
# Calculating the normalized data

data["N_Age"]= pd.Series((data["Age"] - val_min)/(val_max - val_min), index=data.index)
```

```python
# DataFrame with new column

data
```

```python
# Replacing Yes with 1 and No with 0 in Student column

data["Student"]=data["Student"].map({"Yes" : 1 , "No" : 0 , "no" : 0 , "yes" : 1})
```

```python
# Replacing Yes with 1 and No with 0 in Married column

data["Married"]=data["Married"].map({"Yes" : 1 , "No" : 0 , "no" : 0 , "yes" : 1})
```

```python
# Give number of samples in which there are Males who are not students and their age is greater than 60
# Data with gender male

data5=data[data["Gender"]=="Male"]
```

```python
#Checking the missing values in Student and Married column

data_student=data[data["Student"].isna()]
data_student
```

```python
# Replacing missing values from "NaN" to 0 in Student column

data["Student"].fillna(0 , inplace = True)
```

```python
data
```

```python
# Values in column 'Income' are in the range of 10^4 while the values in column "Limit" in thousands. 
# Create a derived column from these two columns

data["Income"]=data["Income"].astype(str)
data["Income"]=data.Income.apply(lambda x : x[:-1] if x[-1]=='%' else x)
data["Income"]=data.Income.apply(lambda x : x[1:] if x[0]=='$' else x)
data["Income"]=data.Income.apply(lambda x : x[:-2]+x[-1] if x[-2]=='&' else x)
data["Income"]=data["Income"].astype(float)
data["Derived_Credit_Balance"] = pd.Series(data["Limit"] / data["Income"] , index=data.index)
data
```

```python
# Age_Group Column
# 0 < Age < 40 Age_Group = 1
# 41 < Age < 60 Age_Group = 2
# 61 < Age < 80 Age_Group = 3
# 81 < Age < 100 Age_Group = 4

data=data.assign(Age_group=pd.cut(data['Age'], bins=[0, 40, 60, 80, 100], right=False, labels=[1, 2, 3, 4]))
```
