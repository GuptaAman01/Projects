<center><h2>Neural Networks Implementation on Logic Gates</center></h2>

```python
import numpy as np    # For statistical calculations
import matplotlib.pyplot as plt   # For plotting graphs
```

```python
feature_set = np.array([[0,0],[0,1],[1,0],[1,1]])  # Inputs
print(feature_set.shape)   # Checking the shape of the array
feature_set   # Printing the array
```

<center><h2>OR Gate</center></h2>

```python
target_output = np.array([[0,1,1,1]])     # Targeted Output
target_output = target_output.reshape(4,1)   # Reshaping the output
print(target_output.shape)   # Checking the shape of the array
target_output   # Printing the target output
```

```python
weights = np.array([[0.1],[0.2]])   # Weights 
print(weights.shape)  # Shape of the weight
weights  # Printing the weights
```

```python
bias = 0.3   # bias unit
lr = 0.05   # Learning rates
```

```python
# Sigmoid function
# sigma = 1/(1+e^(-x))
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
# Derivative of the sigmoid function
def sigmoid_der(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
# The main logic for predicting output and updating the weight values:

for epoch in range(10000):
    inputs = feature_set

    # feedforward input
    in_o = np.dot(feature_set, weights) + bias

    #feedforward output
    out_o = sigmoid(in_o)


    # backpropagation calculating error
    error = out_o - target_output

    # applying formula
    print(error.sum())

    # Calculating derivative
    derror_douto = error
    douto_dino = sigmoid_der(out_o)

    # Multiplying individual derivatives
    deriv = derror_douto * douto_dino

    # Multiplying with the 3rd individual derivative
    # finding the transpose of input features
    inputs = feature_set.T
    deriv_final=np.dot(inputs, deriv)
    
    #updating the weight values
    weights -= lr * deriv_final

    # updating the bias weight value
    for i in deriv:
        bias -= lr * i
```

```python
print(weights)  # Printing new values of weights
print(bias)  # Printing new values of bias units
```

```python
# Predictiong for (1,0)

#taking input
pt=np.array([1,0])

# 1st step
result1= np.dot(pt,weights)+bias

# 2nd step
result2=sigmoid(result1)
    
#print result
print(result2.round())
```

```python
# Predictiong for (1,1)

#taking input
pt=np.array([1,1])

# 1st step
result1= np.dot(pt,weights)+bias

# 2nd step
result2=sigmoid(result1)
    
#print result
print(result2.round())
```

## CONCLUSION:
* The target value for (1,1) is 1 according to the Truth Table of OR Gate, 
  and the predicted output is also 1.
* The target value for (1,0) is 1 according to the Truth Table of OR Gate,
  and the predicted output is also 1.
